```python
import requests
from time import *
import statistics
import matplotlib.pyplot as plt
# Введем функцию для получения данных
def calc(y, theta):
  data = []
  for i in range(1, 50): #делаем много запросов
    url = 'http://dev.astroedu.ru:8000/blackbox?y=' + str(y) + '&theta=' + str(theta)
    r = requests.get(url)
    if str(r.text.encode('utf8')).split('\'')[1] == '"no image"':
      continue
    else:
      data.append(float(str(r.text.encode('utf8')).split('\'')[1]))
    return sum(data)/len(data) # среднее по множеству итераций
```

```python
xs = []
ys = []
for ye in range(-40, 40):
    c = calc(ye, 0)
    if not(c == None):
        xs.append(ye)
        ys.append(c)
plt.plot(xs, ys)
plt.xlabel('y, мм')
plt.ylabel('y\', мм')
plt.savefig('prism.png')
plt.show()
# Из массива, который можно запросить, длина призмы - 49 мм
```
